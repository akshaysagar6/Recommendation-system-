```python
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules 
```

```python
new_df=pd.get_dummies(df)
new_df
```

```python
frequent_items=apriori(new_df,min_support=0.15,use_colnames=True)
frequent_items
```

```python
rules=association_rules(frequent_items,metric='lift')
rules.sort_values(by='conviction',ascending=False)[:10]
```

### Higher the conviction excellend rule

### leverage=-1--->negative correration,0--> nocorrelation,1>positive correlation
