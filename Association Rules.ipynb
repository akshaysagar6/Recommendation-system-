```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori,association_rules
import pandas as pd
import numpy as np
```

```python
itemlist=[['milk','onion','nutmeg','kidney beans','eggs','yogurt'],
         ['dill','onion','nutmeg','kidney beans','eggs','yogurt'],
         ['milk','apple','kidney beans','eggs'],
         ['milk','onion','corn','kidney beans','yogurt'],
         ['corn','onion','onion','kidney beans','ice cream','eggs']]
```

```python
te=TransactionEncoder()
te_arr=te.fit_transform(itemlist)
te_arr
```

```python
te.columns_
```

```python
df=pd.DataFrame(data=te_arr,columns=te.columns_)
```

```python
df
```

```python
frequent_items=apriori(df,min_support=.6,use_colnames=True)
```

```python
frequent_items
```

```python
asso_rules=association_rules(frequent_items,min_threshold=1.0)#confidence=1.0
```

```python
asso_rules
```

```python
asso_rules[asso_rules['lift']>1]
```
