```python
user_item_table=movies.pivot_table(index='userId',values='rating',columns='movie').reset_index()
```

```python
user_item_table.index=movies.userId.unique()
```

```python
user_item_table.isnull().sum()
```

```python
user_item_table.fillna(0,inplace=True)
```

```python
user_item_table
```

```python
from sklearn.metrics import pairwise_distances
from sklearn.metrics.pairwise import cosine_similarity
```

```python
user_sim=1 - pairwise_distances(user_item_table.values,metric='cosine')
```

```python
np.fill_diagonal(user_sim,0)
```

```python
user_sim_df=pd.DataFrame(user_sim,index=movies['userId'].unique(),columns=movies['userId'].unique())
```
