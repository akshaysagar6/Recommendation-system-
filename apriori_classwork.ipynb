```python
import pandas as pd
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
items_list = [['Milk', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
           ['Dill', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
           ['Milk', 'Apple', 'Kidney Beans', 'Eggs'],
           ['Milk', 'Onion', 'Corn', 'Kidney Beans', 'Yogurt'],
           ['Corn', 'Onion', 'Onion', 'Kidney Beans', 'Ice cream', 'Eggs']]
```

```python
te = TransactionEncoder()
te_arr = te.fit_transform(items_list)
te_arr
```

```python
te.columns_
```

```python
df = pd.DataFrame(data=te_arr,columns=te.columns_)
df
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
freq_items = apriori(df, min_support=0.6, use_colnames=True)
```

```python
freq_items
```

```python
assoc_rules = association_rules(freq_items, min_threshold=1.0)
assoc_rules
```

```python
assoc_rules[assoc_rules['lift'] > 1]
```

```python
assoc_rules.sort_values(by='lift', ascending=False)[:3]
```
