```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
for i in df:
    sns.countplot(x=i,data=df)
    plt.show()
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
```

### MIN SUPPORT =10% AND threshold=50%

```python
frequency=apriori(df,min_support=0.1,use_colnames=True)
```

```python
frequency
```

```python
len(frequency)
```

```python
frequency.sort_values("support", ascending=True)
```

```python
association=association_rules(frequency,min_threshold=0.5,metric='lift')
```

```python
association
```

```python
one=association[(association['lift']>=1) & (association['confidence'] >=0.7)]
one
```

```python
plt.plot(one.support,label='support')
plt.plot(one.confidence,label='confidence')
plt.legend()
```

```python
plt.plot(one.lift,label='lift')
plt.plot(one.support,label='support')
plt.legend()
```

```python
plt.figure(figsize=(6,6))
sns.scatterplot(x=one['support'],y=one['confidence'])
```

```python
plt.figure(figsize=(6,6))
sns.scatterplot(x=one['support'],y=one['lift'])
```
