```python
from mlxtend.frequent_patterns import association_rules,apriori
```

### SETTING SUPPORT=10% AND THRESHOLD=50%

```python
rule1=apriori(df_new,min_support=0.1,use_colnames=True)
```

```python
rule1
```

```python
rule_1=association_rules(rule1,min_threshold=0.5,metric='lift')
rule_1
```

```python
rule_2=rule_1[(rule_1.lift<2) & (rule_1.confidence>=.4)]
```

```python
rule_2
```

```python
plt.plot(rule_2.support,label='support')
plt.plot(rule_2.confidence,label='confidence')
plt.legend()
```

```python
sns.scatterplot(x=rule_2.support,y=rule_2.confidence)
```

### setting support and threshold same as above but giving max_length for apriori

```python
rule2=apriori(df_new,min_support=.2,use_colnames=True,max_len=2)
```

```python
rule2
```

```python
rule2_1=association_rules(rule2,min_threshold=0.5,metric='lift')
```

```python
rule2_1
```

```python
rule2_2=rule2_1[(rule2_1.lift<4)]
```

```python
plt.plot(rule2_2.lift,label='lift')
plt.plot(rule2_2.confidence,label='confidence')
plt.legend()
```

```python
plt.plot(rule2_2.support,label='support')
plt.plot(rule2_2.confidence,label='confidence')
plt.legend()
```

```python
sns.scatterplot(x=rule2_2.support,y=rule2_2.confidence)
```
